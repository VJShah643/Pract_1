```python
import numpy as np
import pandas as pd
df = pd.read_csv("kc_house_data.csv")
```

```python
print(df)
```

```python
mean = np.mean(df['sqft_living'])
median = np.median(df['sqft_living'])
std_dev = np.std(df['sqft_living'])
print(f"Sqft living : \nMean = {mean} \nMedian = {median} \nStandard deviation = {std_dev}")
```
